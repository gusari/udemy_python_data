```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect("sakila.db")
```

```python
def sql_to_df(sql_query):
    df=pd.read_sql(sql_query,con)
    return df
```

```python
query = '''select first_name,last_name 
                  from customer'''
sql_to_df(query).head()
```

```python
query = '''select country_id from city'''
sql_to_df(query).head()
```

```python
query = '''select * from customer where store_id=1'''
sql_to_df(query).head()
```

```python
query='''select * from customer where first_name= 'MARY' '''
sql_to_df(query).head()
```

```python
query='''select * from film 
                where release_year=2006
                and rating = 'R' '''
sql_to_df(query)
```

```python
query = ''' SELECT *
            FROM film
            WHERE rating = 'PG'
            OR rating = 'R' '''

sql_to_df(query).head()
```

```python
query='''select count(customer_id)
                from customer'''
sql_to_df(query)
```

```python
query='''select * 
                from customer
                 where first_name like 'M%' '''
sql_to_df(query)
```

```python
query='''select * from customer
               where last_name like '_ING' '''
sql_to_df(query)
```

```python
query=''' select * from customer
                where first_name GLOB '[AB]*' '''
sql_to_df(query)
```

```python
query='''select * from customer order by last_name'''
sql_to_df(query)
```

```python
query='''select * from customer order by last_name desc'''
sql_to_df(query)
```

```python
query = '''select store_id, count(customer_id)
                  from customer
                  group by store_id'''
sql_to_df(query)
```
